# 5. What day of the week is the least active?

```python
dt['week']=['week1' if x<=7 else 'week2' for x in dt['Day']]
dt.iloc[200]
```

```python
dt['Day_of_Week'] = pd.to_datetime(dt['Trip date and time'], format='%d-%m-%Y %H:%M').dt.weekday_name 
dt.iloc[2001]
```

```python
least_week1=dt[['week','Day_of_Week','Trip ID']]
def check(week):
    least_week=least_week1.loc[(least_week1['week']==week)]
    least_week=least_week.groupby(['week','Day_of_Week']).count()['Trip ID'].nsmallest(1)  
    print(least_week)   
check("week1")
check("week2")
```

# 8. For a Jeddah Economy driver, what is the expected change in his daily income if we give this driver a Business car instead?

# 10. How much would GoCarzz lose if it rounds down (to the nearest integer)?

```python
import numpy as np

flr=dt[['Trip price (in local currency)','driver earnings (in local currency)','Trip ID']]
flr['Trip floor']=flr['Trip price (in local currency)'].apply(np.floor)
flr['Trip floor']
x=flr['Trip price (in local currency)']-flr['driver earnings (in local currency)']
x=x.sum()
y=flr['Trip floor']-flr['driver earnings (in local currency)']
y=y.sum()
loss_Gocar=x-y
print("Gocarzz loss")
loss_Gocar
```

# 11. What is GoCarzz owed at the end of Oct 7, 2015? (Assume cash is collected every Sunday)

```python
owed_4=dt[['Day','Trip price (in local currency)','driver earnings (in local currency)','Trip ID' ]]
owed_4['Gocarzz_pro']=owed_4['Trip price (in local currency)']-owed_4['driver earnings (in local currency)']

owed_4['Day']=owed_4.loc[(owed_4['Day']<=4)]
owed_4=owed_4.dropna()   
total_4_days=owed_4['Gocarzz_pro'].sum()
owed_7=dt[['Day','Trip price (in local currency)','driver earnings (in local currency)','Trip ID' ]]
owed_7['Gocarzz_pro']=owed_7['Trip price (in local currency)']-owed_7['driver earnings (in local currency)']
owed_7['Day']=owed_7.loc[(owed_7['Day']<=7)]
owed_7=owed_7.dropna()
total_7_days=owed_7['Gocarzz_pro'].sum()
print("GoCarzz owed at the end of Oct 7, 2015")
Owed_Amt=total_7_days-total_4_days
Owed_Amt
```
